```python
import pandas as pd
import textstat
from tqdm import tqdm

textstat.set_lang('en')
```

```python
json_filename = "all_utterances.json"
data_df = pd.read_json(json_filename)
```

```python
kf_total = []
spache_total = []

for file in tqdm(range(len(data_df))):
    kf_ease = []
    spache_ease = []
    for sentence in data_df['plain_utterances'][file]:
        clean_sentence = sentence.replace("POSTCLITIC", "")
        kf_ease.append(textstat.flesch_reading_ease(clean_sentence))
        spache_ease.append(textstat.spache_readability(clean_sentence))
    kf_total.append(kf_ease)
    spache_total.append(spache_ease)
```

```python
utterance_strs = [" ".join(u) if isinstance(u, (list, tuple)) else (u or "") for u in data_df['plain_utterances']]
data_df['plain_utterances_str'] = utterance_strs
```

```python
data_df['kf_ease'] = kf_total
data_df['spache_ease'] = spache_total
data_df['kf_ease_avg'] = [sum(x)/len(x) for x in kf_total]
data_df['kf_ease_total_avg'] = [textstat.flesch_reading_ease(x) for x in data_df['plain_utterances_str']]
data_df['spache_avg'] = [sum(x)/len(x) for x in spache_total]
data_df['spache_ease_total_avg'] = [textstat.spache_readability(x) for x in data_df['plain_utterances_str']]
```

```python
order = ['dataset', 'filename', 'kf_ease_avg', 'spache_avg', 'kf_ease', 'spache_ease', 'kf_ease_total_avg',  'spache_ease_total_avg', 'plain_utterances', 'plain_utterances_str']
data_df = data_df[order]
```
