```python
import pandas as pd
import textstat
from tqdm import tqdm

textstat.set_lang('en')
```

```python
csv_filename = "all_utterances.json"
data_df = pd.read_json(csv_filename)
```

```python
data_df['plain_utterances'][102]
```

```python
kf_total = []
spache_total = []

for file in tqdm(range(len(data_df))):
    kf_ease = []
    spache_ease = []
    sentences = []
    for sentence in data_df['plain_utterances'][file]:
        clean_sentence = sentence.replace("POSTCLITIC", "")
        sentences.append(clean_sentence)
        kf_ease.append(textstat.flesch_reading_ease(clean_sentence))
        spache_ease.append(textstat.spache_readability(clean_sentence))

    data_df['plain_utterances'][file] = sentences
    
    kf_total.append(kf_ease)
    spache_total.append(spache_ease)
```

```python
data_df['kf_ease'] = kf_total
data_df['spache_ease'] = spache_total
data_df['kf_ease_avg'] = [sum(x)/len(x) for x in kf_total]
data_df['spache_avg'] = [sum(x)/len(x) for x in spache_total]
```

```python
data_df.to_csv('Score_df.csv', index=False)
```
